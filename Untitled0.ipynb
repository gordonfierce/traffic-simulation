```python
def run_multiple_simulations():
    num_simulations = 100
    time = 60
    mean_speeds = np.zeros((num_simulations, time))
    for number in range(num_simulations):
        road = Road()
        simulation = Simulation(road)
        simulation.run_simulation()
        mean_speeds[number] += simulation.speeds.mean(axis=1)
        
    return mean_speeds
```

```python
many_simulation_results = run_multiple_simulations()
```

```python
plt.plot(many_simulation_results.mean(axis=0))
```
